# 1.3 Perform Univariate analysis - Plot PDF, CDF, Boxplot, Voilin plots

## PDF and CDF

```python
data_survival_status_1 = data[data['survival_status']==1]
data_survival_status_2 = data[data['survival_status']==2]
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10,7))
plt.subplots_adjust(top=2,bottom=0.8)

counts, bin_edges = np.histogram(data_survival_status_1['age'],
                                 density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)

ax[0].plot(bin_edges[1:],pdf, label="Probability Density Function")
ax[0].plot(bin_edges[1:], cdf,label="Cummalative Density Function")
ax[0].set_xlim([30,90])
ax[0].set_xlabel('Age')
ax[0].set_ylabel('Probability')
ax[0].legend(loc='upper left')
ax[0].title.set_text("Survived")
counts, bin_edges = np.histogram(data_survival_status_2['age'],
                                 density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
ax[1].plot(bin_edges[1:],pdf, label="Probability Density Function")
ax[1].plot(bin_edges[1:], cdf,label="Cummalative Density Function")
ax[1].set_xlim([30,90])
ax[1].set_xlabel('Age')
ax[1].set_ylabel('Probability')
ax[1].title.set_text("Not Survived")
ax[1].legend(loc='upper left')
plt.show()
```

#### Age less than 40 have more chances of survival
#### Age more than 76 have less chances of survival 

```python
fig, ax = plt.subplots(2, 1, figsize=(10,7))
plt.subplots_adjust(top=2,bottom=0.8)
counts, bin_edges = np.histogram(data_survival_status_1['operation_Year'],
                                 density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)

ax[0].plot(bin_edges[1:],pdf, label="Probability Density Function")
ax[0].plot(bin_edges[1:], cdf,label="Cummalative Density Function")
ax[0].set_xlim([55,70])
ax[0].set_xlabel('operation_Year')
ax[0].set_ylabel('Probability')
ax[0].legend(loc='upper left')
ax[0].title.set_text("Survived")
counts, bin_edges = np.histogram(data_survival_status_2['operation_Year'],
                                 density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
ax[1].plot(bin_edges[1:],pdf, label="Probability Density Function")
ax[1].plot(bin_edges[1:], cdf,label="Cummalative Density Function")
ax[1].set_xlim([55,70])
ax[1].set_xlabel('operation_Year')
ax[1].set_ylabel('Probability')
ax[1].legend(loc='upper left')
ax[1].title.set_text("Not Survived")
plt.show()
```
