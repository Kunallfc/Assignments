```python
def replace_digits(A):
    encoded_list = []
    for index in range(len(A)):
        if (A[index]).isnumeric():
            encoded_list.append("#")
        else:
            encoded_list.append(A[index])

    return "".join(encoded_list)
```

```python
import math


def closest_points_to_p(S, P):
    points_distance = {}
    for each in S:
        points_distance[each] = cosine_distance(each[0], each[1], P[0], P[1])
    closest_points_to_p = [i[0] for i in sorted(points_distance.items(), key=lambda x: x[1])[:5]]
    return closest_points_to_p  # its list of tuples


def cosine_distance(x, y, px, py):
    return math.acos((x * px + y * py) / (math.sqrt(x ** 2 + y ** 2) * math.sqrt(px ** 2 + py ** 2)))


S = [(1, 2), (3, 4), (-1, 1), (6, -7), (0, 6), (-5, -8), (-1, -1), (6, 0), (1, -1)]
P = (3, -4)
points = closest_points_to_p(S, P)
print(points)  # print the returned values
```
